```python
class Coordinate(object):
    """A coordinate made up of an x and y values."""
    def __init__(self, x, y):
   # initlizae a coordinate with x and y values
        self.x = x
        self.y = y

    def distance(self, other_coordinate):
        """Compute the distance between this coordinate and another."""
        dx = self.x - other_coordinate.x
        dy = self.y - other_coordinate.y
        return (dx ** 2 + dy ** 2) ** 0.5
```

```python
c= Coordinate(3,4)
print(c.x)
```

```python
origin = Coordinate(0,0)
print(origin.x, origin.y)
#1st way to call distance method best
dist = c.distance(origin)
print(dist)
#2nd way to call distance method
dist = Coordinate.distance(c, origin)
print(dist)
#3rd way to call distance method
origin.distance(c)
```

```python
import numpy as np

def euclidean_distance(point1, point2):
    """Compute the Euclidean distance between two points."""
    point1 = np.array(point1)
    point2 = np.array(point2)
    return  np.sqrt(np.sum((point1-point2)**2))
```

```python
#example 
point1 = np.array([1, 2, 3])
point2 = np.array([4, 5, 6])
dist = euclidean_distance(point1, point2)
print(dist)
```
