```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#LOdaing the image 
image = Image.open('Moon_Farside_LRO.jpg').convert('L') # Convert to grayscale

#convert to floatng point array 
A = np.array(image, dtype = np.float64)

print(A.shape)
```

```python
U, S, Vt = np.linalg.svd(A, full_matrices=True)

#construct sigma as a matrix 
Sigma = np.zeros((U.shape[1], Vt.shape[0]))
np.fill_diagonal(Sigma, S)
```

```python
#verifyinh the orthogonality of U and Vt
#U^T * U -I and V^t * V -I 

#identity matrixes 
I_U = np.eye(U.shape[1])
I_V = np.eye(Vt.shape[0])

#frobenius norm of the difference between U^T * U and I_U
epsilon_U = np.linalg.norm(U.T @ U - I_U, ord = 'fro')

#frobenius norm of the difference between V^T * V and I_V
epsilon_V = np.linalg.norm(Vt.T @ Vt - I_V, ord = 'fro')

print(f"Error in U orthogonality: {epsilon_U}")
print(f"Error in V orthogonality: {epsilon_V}")

# Optional: show image
plt.imshow(A, cmap="gray")
plt.title("Original Grayscale Image")
plt.axis("off")
plt.show()
```
