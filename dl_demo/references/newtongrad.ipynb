```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#constant 
c= 0.15 * np.pi

#function and derivatives 
def f(x):
    return x*np.cos(c*x)
def df(x):
    return np.cos(c*x) - c*x*np.sin(c*x)    
def ddf(x):
    return -2*c*np.sin(c*x) - c**2*x*np.cos(c*x)

#gradient descent 
def grad_descent(x0, alpha=0.1, max_iter=100):
    
    trace = [x0]
    x = x0

    for i in range(max_iter):
        x_new = x - alpha * df(x)
        trace.append(x)
        if abs(df(x)) < 1e-6 or abs(df(x_new)) < 1e-6:
            break
        x = x_new

    return np.array(trace)

#newton's method
def newton_method(x0, max_iter=100):
    
    trace = [x0]
    x = x0

    for i in range(max_iter):
        dfx = df(x)
        ddfx = ddf(x)
        if abs(dfx) < 1e-6 or abs(ddfx) < 1e-6:
            break
        x_new = x - dfx / ddfx
        trace.append(x)
        x = x_new

    return np.array(trace)
```

```python
x0_list = [6.2, -6.1]

#plot for gradient descent
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
x = np.linspace(-10, 10, 400)
plt.plot(x, f(x), label='f(x)')
for x0 in x0_list:
    trace = grad_descent(x0)
    plt.plot(trace, f(trace), marker='o', label=f'Grad Descent (x0={x0})')
plt.title('Gradient Descent')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
#plot for newton's method
plt.subplot(1, 2, 2)
plt.plot(x, f(x), label='f(x)')
for x0 in x0_list:
    trace = newton_method(x0)
    plt.plot(trace, f(trace), marker='o', label=f'Newton\'s Method (x0={x0})')
plt.title('Newton\'s Method')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.tight_layout()
plt.show()
```

Speed of Convergence:
Newton’s method converges quadratically and is much faster near the solution, while gradient descent converges linearly and is slower.

Sensitivity to Initialization:
Gradient descent is less sensitive and always moves in the direction of decreasing function value. Newton’s method is highly sensitive to the starting point and may converge to a local maximum or saddle point instead of a minimum.
